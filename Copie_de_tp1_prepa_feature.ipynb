## 3. Traitement des données manquantes (à faire sur les données Train et Test)

3.1 La variable "Embarked" possède des valeurs nulles, compléter par la valeur majoritaire (S).

```python
train['Embarked'].fillna('S',inplace=True)
test['Embarked'].fillna('S',inplace=True)
```

3.2 Afficher l'influence du port d'embarquement sur la Survie 

```python
train[['Survived','Embarked']].groupby(['Embarked']).mean()*100
```

3.3 La variable "Fare" (prix du ticket) possède des valeurs manquantes. Remplacer ces valeurs par la valeur médiane.

```python
train['Fare'].fillna(train['Fare'].median(),inplace=True)
test['Fare'].fillna(test['Fare'].median(),inplace=True)
```

3.4 Créer une variable CategoricalFare Découpant la valeur de ticket en 4 catégories possédant des nombres de passagers à peu près égaux. 

```python
train['CategoricalFare'] = pd.qcut(train['Fare'],4) 
test['CategoricalFare'] = pd.qcut(test['Fare'],4) 
```

3.5 Analyser l'effet de ces catégories de prix de ticket sur la survie.

```python
train[['Survived','CategoricalFare']].groupby(['CategoricalFare']).mean() * 100
```

3.6 Combien de valeurs d'Age manquent ? Complète par des valeurs entières aléatoires entre (moyenne - écart type) et (moyenne + écart type).

```python
#train['Age'].isna().sum() #= 177
train['Age'].fillna(np.random.randint(train['Age'].mean() - train['Age'].std(),train['Age'].mean() + train['Age'].std()),inplace=True)
test['Age'].fillna(np.random.randint(test['Age'].mean() - test['Age'].std(),test['Age'].mean() + test['Age'].std()),inplace=True)
```

3.7 Créer une variable CategoricalAge Découpant la valeur d'age en 5 catégories égales. 

```python
train['CategoricalAge'] = pd.qcut(train['Age'],5) 
test['CategoricalAge'] = pd.qcut(test['Age'],5)
train
```

3.8 Analyser l'effet de ces catégories d'age sur la survie.

```python
train[['Survived','CategoricalAge']].groupby(['CategoricalAge']).mean() * 100
```

## 4 : Traitement de features complexes

4.1 : Le nom permet de retrouver le titre du passager. Trouver les titres lorsqu'ils existent.

```python
train['Title']=train.Name.apply(lambda st: st[st.find(" ")+1:st.find(".")])
test['Title'] = test.Name.apply(lambda st: st[st.find(" ")+1:st.find(".")])
test
```

4.2 : En utilisant la fonction pandas crosstab, afficher pour train le nombre d'occurences par titres (en ligne) et par sexe (en colonne)

```python
pd.crosstab(train.Title,train.Sex)
```

4.3 : 
Remplacer 'Melle et 'Ms' par 'Miss'
Remplacer 'Mme' par 'Mrs'
Creer une valeur "Rare" pour les autres titres avec moins de 10 occurences. 

```python
train['Title'] = train['Title'].str.replace('Mlle|Ms','Miss')
train['Title'] = train['Title'].str.replace('Mme','Mrs')
test['Title'] = test['Title'].str.replace('Mlle|Ms','Miss')
test['Title'] = test['Title'].str.replace('Mme','Mrs')

valuesTrain = train.Title.value_counts()
toChangeTrain = train.Title.isin(valuesTrain.index[valuesTrain<10])
train.loc[toChangeTrain, 'Title'] = 'Rare'

valuesTest = test.Title.value_counts()
toChangeTest = test.Title.isin(valuesTest.index[valuesTest<10])
test.loc[toChangeTest, 'Title'] = 'Rare'
test
```
